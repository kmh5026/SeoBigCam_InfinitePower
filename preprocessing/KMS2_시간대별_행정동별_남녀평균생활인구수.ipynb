## 1. 서울시 생활인구(2021년 7월) 데이터 불러오기

```python
import numpy as np 
import pandas as pd 
```

```python
dt = pd.read_csv('LOCAL_PEOPLE_DONG_202107.csv')
dt.head()
```

```python
list(dt.columns)
```

## 2. 남녀 25~64세 생활인구 수

### 남자 25~64세 인구 데이터만 추출

```python
dt_male=dt[list(dt.columns)[8:16]]
dt_male.head()
```

```python
dt['남자25세부터64세생활인구수']=dt_male.sum(axis=1)
dt.head()
```

### 여자 25~64세 인구 데이터만 추출

```python
dt_female=dt[list(dt.columns)[22:30]]
dt_female.head()
```

```python
dt['여자25세부터64세생활인구수']=dt_female.sum(axis=1)
dt.head()
```

## 3. 기준일, 시간, 행정동코드, 25세~64세 인구 수로 구성된 데이터 프레임 생성

```python
col=['기준일ID', '시간대구분', '행정동코드', '남자25세부터64세생활인구수', '여자25세부터64세생활인구수']
```

```python
dt=dt[col]
dt.head()
```

## 4. 행정동코드 별 7_9(출근), 12_14(점심), 18_20(퇴근) 시간대 인구 수

```python
code=dt['행정동코드'].unique()
code
```

```python
len(code)
```

### 7시~9시 행정동(코드)별 남녀평균생활인구 수

```python
dt7_9 = dt[(dt['시간대구분']>6)&(dt['시간대구분']<10)]
dt7_9
```

```python
dt7_9 = dt7_9.groupby('행정동코드').mean()
dt7_9.reset_index(drop=False, inplace=True)
dt7_9.drop('기준일ID' ,axis=1, inplace=True)
dt7_9['시간대구분']='7시~9시'
```

```python
dt7_9
```

### 12시~14시 행정동(코드)별 남녀평균생활인구 수

```python
dt12_14 = dt[(dt['시간대구분']>11)&(dt['시간대구분']<15)]
dt12_14
```

```python
dt12_14 = dt12_14.groupby('행정동코드').mean()
dt12_14.reset_index(drop=False, inplace=True)
dt12_14.drop('기준일ID' ,axis=1, inplace=True)
dt12_14['시간대구분']='12시~14시'
```

```python
dt12_14
```

### 18시~20시 행정동(코드)별 남녀평균생활인구 수

```python
dt18_20 = dt[(dt['시간대구분']>17)&(dt['시간대구분']<21)]
dt18_20
```

```python
dt18_20 = dt18_20.groupby('행정동코드').mean()
dt18_20.reset_index(drop=False, inplace=True)
dt18_20.drop('기준일ID' ,axis=1, inplace=True)
dt18_20['시간대구분']='18시~20시'
```

```python
dt18_20
```

## 5. 행정동코드와 행정동 매핑

### '행정_법정_매칭_new.txt' 데이터 불러오기

```python
name=pd.read_table('행정_법정_매칭_new.txt',sep=',')
name.head()
```

```python
name=name.drop(['법정동','법정동코드'], axis=1)
name.drop_duplicates(subset=None, keep='first', inplace=True)
name.reset_index(drop=True, inplace=True)
name.head()
```

### 행정동코드를 행정동명으로 바꾸는 dictionary 생성

```python
Dic=dict(zip(name['행정동코드'], name['행정동']))
Dic
```

### 데이터프레임에 행정동 열 추가

```python
for i in range(424):
    dt7_9.loc[i,'행정동']=Dic.get(dt7_9.loc[i,'행정동코드']*100)
    dt12_14.loc[i,'행정동']=Dic.get(dt12_14.loc[i,'행정동코드']*100)
    dt18_20.loc[i,'행정동']=Dic.get(dt18_20.loc[i,'행정동코드']*100)
```
