## 1. 서울시_우리마을가게_상권분석서비스(상권-점포)_2020년_4분기 데이터 불러오기

```python
import numpy as np
import pandas as pd
```

```python
market=pd.read_csv("서울시_우리마을가게_상권분석서비스(상권-점포)_2020년.csv")
```

```python
market.head()
```

### 4분기 데이터만 추출

```python
market=market[market['기준_분기_코드']==4]
```

### NaN(상권구분코드 K )삭제

```python
idx=market[market['상권_구분_코드']=='K'].index
```

```python
market=market.drop(idx)
```

```python
market['상권_구분_코드'].unique()
```

### 필요없는 열 삭제

```python
market = market.drop(list(market.columns)[10:], axis=1)
market = market.drop(['상권_구분_코드', '상권_구분_코드_명', '상권_코드_명', '서비스_업종_코드', '서비스_업종_코드_명'], axis=1)
market.head()
```

## 2. 상권_코드(주소)별 점포수 집계

### 상권 코드 공백제거

```python
market['상권_코드'].nunique()
```

```python
market['상권_코드'] = market['상권_코드'].astype(int)
```

```python
market['상권_코드'].nunique()
```

```python
market_gr = market.groupby(market['상권_코드']).sum()
market_gr.head()
```

```python
market_gr.shape
```

### 연도, 분기 수정

```python
market_gr['기준_년_코드'] = 2020
market_gr['기준_분기_코드'] = 4
market_gr = market_gr.reset_index()
```

```python
market_gr.head()
```

## 3. 상권코드를 행정동 코드로 변환

### 상권영역 데이터 불러오기

```python
area=pd.read_csv('서울시 우리마을가게 상권분석서비스(상권영역).csv')
```

```python
area.head()
```

```python
area['상권_코드'].nunique()
```

### 상권코드를 행정동 코드로 바꾸는 dictionary 생성

```python
Dic=dict(zip(area['상권_코드'], area['행정동_코드']))
```

```python
Dic
```

## 4. market_gr에 행정동_코드 열 추가

```python
market_gr.head()
```

```python
for i in range(1496):
    market_gr.loc[i,'행정동_코드']=Dic[market_gr.loc[i,'상권_코드']]
```

```python
market_gr.head()
```

```python
market_gr['행정동_코드'] = market_gr['행정동_코드']*100
market_gr['행정동_코드'] = market_gr['행정동_코드'].astype(int)
market_gr.head()
```

## 5. '행정동_코드'를 '행정동명' 으로 변경

### 행정_법정_매칭_new.txt 데이터 불러오기

```python
name=pd.read_table('행정_법정_매칭_new.txt',sep=',')
name.head()
```

```python
name=name.drop(['법정동','법정동코드'], axis=1)
name.drop_duplicates(subset=None, keep='first', inplace=True)
name.head()
```

```python
name.reset_index(drop=True, inplace=True)
name.head()
```

### 행정동코드를 행정동명으로 바꾸는 dictionary 생성

```python
Dic2=dict(zip(name['행정동코드'], name['행정동']))
```

```python
Dic2
```

### market_gr에 행정동명 추가

```python
market_gr.head()
```

```python
for i in range(1496):
    market_gr.loc[i,'행정동']=Dic2.get(market_gr.loc[i,'행정동_코드'])
```
