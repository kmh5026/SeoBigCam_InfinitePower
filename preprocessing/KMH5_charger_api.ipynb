```python
data = pd.read_table('./data/api_charger.txt', sep=',')

from itertools import repeat
data['cnt'] = list(repeat(1, len(data)))

print(data.shape)
data.head()
```

```python
dt = data.groupby(['statNm', 'statId', 'addr', 'lat', 'lng', 'bnm', 'busiNm'])
gr = dt.sum('cnt')
gr = gr.reset_index()

print(gr.shape)
gr.head()
```

```python
gr[gr['addr']=='서울특별시 서초구 신반포로 270']  ### 같은 주소인데 충전소가 뿔뿔이 흩어져 있는 건 어떡하지...
```
