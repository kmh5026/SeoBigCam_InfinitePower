## 2020년 4분기가 가장 깔끔

```python
market = pd.read_csv('./data/서울시_우리마을가게_상권분석서비스(상권-점포)_2020년.csv')

market = market[(market['기준_년_코드']==2020) & (market['기준_분기_코드']==4)]
market = market.reset_index(drop=True)

print(market.shape)
market.head(n=10)
```

```python
market = market[market['상권_구분_코드']!='K']
market['상권_구분_코드'].unique()
```

```python
market = market.drop(list(market.columns)[10:], axis=1)
market = market.drop(['상권_구분_코드', '상권_구분_코드_명', '상권_코드_명', '서비스_업종_코드', '서비스_업종_코드_명'], axis=1)
market['기준_년_코드'] = 2020
market['기준_분기_코드'] = 4
print(market.shape)
market.head()
```

```python
market['상권_코드'] = market['상권_코드'].astype(int)
len(market['상권_코드'].unique())
```

```python
# market = market.apply(lambda x: x.str.strip(), axis = 1) ### 공백 제거 
```

#### 상권코드 별 점포 수 집계 

```python
market_gr = market.groupby('상권_코드').sum()

market_gr['기준_년_코드'] = 2020
market_gr['기준_분기_코드'] = 4
market_gr = market_gr.reset_index()

print(market_gr.shape)
market_gr.head()
```

```python
data = market_gr.drop(['기준_년_코드', '기준_분기_코드'], axis=1)

print(data.shape)
data.head()
```

```python
data['상권_코드'].unique()
```

## 동별차량등록 데이터에 법정동코드 추가 

```python
dt = pd.read_csv('./data/법정동별_차량등록현황_processed.txt')
print(dt.shape)
dt.head()
```

```python
match = pd.read_csv('./data/행정_법정동코드_매칭.txt')
print(match.shape)
match.head()
```

```python
match['행정동코드'].nunique()
```

```python
match['법정동코드'].nunique()
```

### 시도명 시군구명 읍면동명 합쳐서 법정동 차량등록 데이터에 매칭!! 

```python
code = match[['match', '법정동코드', '동리명']]
code.columns = ['사용본거지법정동명','법정동코드', '동리명']
print(code.shape)
code.head()
```

```python
merged = pd.merge(dt, code, how='outer', on='사용본거지법정동명')
print(merged.shape)
merged.head()
```

```python
na = merged[merged['법정동코드'].isna()]
na
```

#### 결측치 : 동별 대수도 미미하고 찾아보니 다 편입/폐기된 동 이름 >>> 제거

```python
merged = merged[merged['법정동코드'].notna()]
merged['법정동코드'] = merged['법정동코드'].astype(int)
print(merged.shape) 
merged.head()
```

```python
merged.to_csv('./data/법정동별_차량등록현황_processed.txt', sep=',', index=False)
```
