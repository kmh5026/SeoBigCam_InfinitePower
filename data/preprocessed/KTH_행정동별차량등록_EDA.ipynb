# 차종별 동별 순위

```python
rank = data.copy()
```

```python
rank['동별 총 대수'] = data['동별 총 대수'].rank(method='min', ascending=False)
rank['휘발유'] = data['휘발유'].rank(method='min', ascending=False)
rank['경유'] = data['경유'].rank(method='min', ascending=False)
rank['LPG'] = data['LPG'].rank(method='min', ascending=False)
rank['전기'] = data['전기'].rank(method='min', ascending=False)
rank['CNG'] = data['CNG'].rank(method='min', ascending=False)
rank['수소'] = data['수소'].rank(method='min', ascending=False)
rank['기타연료'] = data['기타연료'].rank(method='min', ascending=False)
rank['하이브리드'] = data['하이브리드'].rank(method='min', ascending=False)
rank['전하'] = data['전하'].rank(method='min', ascending=False)
```

```python
rank.iloc[:,2:] = rank.iloc[:,2:].astype('int')
rank.head()
```
